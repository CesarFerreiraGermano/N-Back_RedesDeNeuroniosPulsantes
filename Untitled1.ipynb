```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("darkgrid")
```

```python
df_acertos = pd.read_csv('din_acertos.txt', sep=" ", header=None)
df_tipo = pd.read_csv('din_tipo.txt', sep=" ", header=None)
```

```python
df_acertos[99].value_counts()
```

```python
i = 2
acertos = []
erros = []
while i <= 99:
    acertou = df_acertos[i].value_counts()[1]
    errou = df_acertos[i].value_counts()[2]
    acertou_normalizado = (acertou*100 / (acertou+errou)).round()
    errou_normalizado = (errou*100 / (acertou+errou)).round()
    acertos.append(acertou_normalizado)
    erros.append(errou_normalizado)
    i += 1
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,99),acertos)
plt.xlim((0,50))
plt.xlabel('Janelas',fontsize=13)
plt.ylabel('Porcentagem de acertos',fontsize=13)
plt.title('Porcentagem de acertos da rede ao longo das janelas',fontsize=15)
plt.savefig('acertos.png')
```
